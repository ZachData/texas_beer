```python
def null_sum():

    # Check if there are any null values
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    x = missing.head()
    
    return x
    
null_sum()
```

```python
hou_month = df_hou.groupby(['TABC Permit Number','Taxpayer Number','Location Name','Location Address','Obligation End Date']).agg('sum')
```

```python
hou_month.columns.to_list()
```

```python
hou_month = hou_month[['Liquor Receipts','Wine Receipts','Beer Receipts','Total Receipts']]

print(hou_month.shape)

hou_month.head(5)
```

```python
hou_month['Beer %'] = hou_month['Beer Receipts'] / hou_month['Total Receipts']
```

```python
hou_month['Liquor %'] = hou_month['Liquor Receipts'] / hou_month['Total Receipts']
```

```python
hou_month['Wine %'] = hou_month['Wine Receipts'] / hou_month['Total Receipts']
```

```python
hou_beer = hou_month['Beer %'].mean()

print(hou_beer)
```

```python
hou_liq = hou_month['Liquor %'].mean()

print(hou_liq)
```

```python
hou_wine = hou_month['Wine %'].mean()

print(hou_wine)
```

```python
hou_pcts = [hou_beer, hou_liq, hou_wine]

x_hou = ['beer','liquor', 'wine']

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,6))

plt.bar(x_hou, hou_pcts)
plt.title('Houston')
plt.ylabel('% of all sales', labelpad=-20, loc='top', rotation='horizontal', fontsize=10)
```
